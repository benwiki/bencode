## Approach nr. 1
One simple little cell, getting random instructions on where to move. Pretty chaotic, isn't it?  
Also, take a look at that! Little Benke doens't know about `sleep()` so he does magic with the variable `a`?  
Or was I just lazy to import it? XD

```python
from tkinter import *
from random import randint
ablak=Tk()
rajz=Canvas(width=700, height=700, background='white')
rajz.pack()

moves=200
r=2
pos=[350,600]
thedot=rajz.create_oval(pos[0]-r, pos[1]-r, pos[0]+r, pos[1]+r, width=2*r, outline='red')

class Dot:
    def __init__(self):
        self.taken=0
        self.moved = [[randint(-5, 5), randint(-5, 5)] for i in range(moves)]
        self.thedot=rajz.create_oval(pos[0]-r, pos[1]-r, pos[0]+r, pos[1]+r, width=r*2)

    def move(self):
        if pos[0]>0 and pos[0]<700 and pos[1] > 0 and pos[1] < 700:
            rajz.delete(self.thedot)
            x=pos[0]+self.moved[self.taken][0]
            y=pos[1]+self.moved[self.taken][1]
            self.thedot=rajz.create_oval(x-r, y-r, x+r, y+r, width=r)
            pos[0]=x
            pos[1]=y
            self.taken += 1

ablak.update()
dots=Dot()
a=1
for i in range(moves):
    for k in range(1):
        dots.move()
    for k in range(100000):
        a+=1
        a-=1
    ablak.update()
    
ablak.mainloop()
```

```python
from colorsys import hsv_to_rgb

def coloring(a, b):
    components = hsv_to_rgb((a/b)*0.8, 1, 1)
    rgb = (int(comp*255) for comp in components)
    return '#'+''.join(f'{comp:02X}' for comp in rgb)
```
